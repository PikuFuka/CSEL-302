# **1. Import libraries**

```python
import pandas as pd #Data manipulation
import numpy as np #Numerical Operation
import matplotlib.pyplot as plt #Data Visualization
from sklearn.linear_model import LinearRegression #Data modeling
```

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
data.head()
```

```python
plt.scatter(data['Previous_Exam_Scores'], data['Grades'])
plt.xlabel('Independent Variable (X)')
plt.xlabel('Dependent Variable (Y)')
plt.title('HELLO')
plt.show()
```

```python
X = data[['Previous_Exam_Scores']] #Independent Varibles
Y = data[['Grades']] #Dependent Varibles
```

```python
model = LinearRegression()

model.fit(X, Y)
```

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope', slope)
print('Intercept', intercept)
```

```python
plt.scatter(X, Y)
plt.plot(X, model.predict(X), color = 'RED')
plt.xlabel('Independent variable (X)')
plt.xlabel('Dependent variable (Y)')
plt.title('Linear Regression')
plt.show()
```

```python
new_X = [[np.random.randint(0, 100)]]

predicted_Y = model.predict(new_X)

print('Value of X', new_X)
print('Predicted Value of Y', predicted_Y)
```
